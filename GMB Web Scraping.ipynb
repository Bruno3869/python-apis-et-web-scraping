```python
# import module
import requests 
import bs4 
  
# Taking thecity name as an input from the user
city ="Lyon"
  
# Generating the url  
url ="https://www.google.fr/search?q=temperature+a+"+ city
  
# Sending HTTP request 
request_result = requests.get(url)
  
# Pulling HTTP data from internet 
soup = bs4.BeautifulSoup(request_result.text,"html.parser")
  
# Finding temperature in Celsius.
# The temperature is stored inside the class"BNeawe". 
temp = soup.find("div", class_='BNeawe').text 
    
print(temp)
```

```python
# import module
import requests 
import bs4
from time import sleep
from random import randint
  
# Taking thecity name as an input from the user
shops =["Saint+Bonnet+de+Mure", "Bourges", "Paris+13", "Narbonne", "Lons+le+Saunier"]

#scrape elements
# Generating the url
for shop in shops:
    url ="https://www.google.fr/search?q=aviva+cuisines"+ shop
    
    # Sending HTTP request 
    request_result = requests.get(url)
    
    # Pulling HTTP data from internet 
    soup = bs4.BeautifulSoup(request_result.text,"html.parser")
    
    # Finding the note.
    # The note is stored inside the class"BNeawe". 
    note = soup.find("span", class_='oqSTJd').text 
    print(shop.replace('+',' ') + " a une note de : " + note + "/5")
    sleep(randint(1,4))
```

```python
# import module
import requests 
import bs4
import pandas as pd
from time import sleep
from random import randint

# 1. Redéclarer les listes
notes = []
magasins = []
  
# Taking thecity name as an input from the user
shops =["Saint+Bonnet+de+Mure", "Bourges", "Paris+13", "Narbonne", "Lons+le+Saunier"]

#scrape elements
# Generating the url
for shop in shops:
    url ="https://www.google.fr/search?q=aviva+cuisines"+ shop
    
    # Sending HTTP request 
    request_result = requests.get(url)
    
    # Pulling HTTP data from internet 
    soup = bs4.BeautifulSoup(request_result.text,"html.parser")
    
    # Finding the note.
    # The note is stored inside the class"oqSTJd". 
    if soup.find("span", class_='oqSTJd').text is not None:
        #scrape de la note
        note = soup.find("span", class_='oqSTJd').text
        notes.append(note)
        
        #ajout du magasin
        magasins.append(shop.replace('+',' '))
        
        print(shop.replace('+',' ') + " a une note de : " + note + "/5")
    # Pause de la boucle de 8 à 15 secondes    
    sleep(randint(8,15))
```

```python
shop_ratings = pd.DataFrame({
    'magasin': magasins,
    'note': notes,
    })
```
