```python
# import module
import requests 
import bs4 
  
# Taking thecity name as an input from the user
city ="Lyon"
  
# Generating the url  
url ="https://www.google.fr/search?q=temperature+a+"+ city
  
# Sending HTTP request 
request_result = requests.get(url)
  
# Pulling HTTP data from internet 
soup = bs4.BeautifulSoup(request_result.text,"html.parser")
  
# Finding temperature in Celsius.
# The temperature is stored inside the class"BNeawe". 
temp = soup.find("div", class_='BNeawe').text 
    
print(temp)
```

```python
# import module
import requests 
import bs4 
  
# Taking thecity name as an input from the user
shop ="Saint+Bonnet+de+Mure"
  
# Generating the url  
url ="https://www.google.fr/search?q=aviva+cuisines"+ shop
  
# Sending HTTP request 
request_result = requests.get(url)
  
# Pulling HTTP data from internet 
soup = bs4.BeautifulSoup(request_result.text,"html.parser")
  
# Finding the note.
# The note is stored inside the class"BNeawe". 
note = soup.find("span", class_='oqSTJd').text 
    
print(shop.replace('+',' ') + " a une note de : " + note + "/5")
```
