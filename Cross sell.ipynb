## Scraping du Cross-sell PDM

### Test url Scraping

https://www.placedumarche.fr/p/683-salade-de-fruits-exotiques-1-kg

```python
# conda install -c anaconda beautifulsoup4 à faire avant dans la console
import requests
# Librairie BeautifulSoup du package bs4
from bs4 import BeautifulSoup
```

```python
# Télécharger la page
from requests import get
url = 'https://www.placedumarche.fr/p/683-salade-de-fruits-exotiques-1-kg'
response = requests.get(url)
print(response.text)
```

### Structure HTML de la page

```python
from bs4 import BeautifulSoup
html_soup = BeautifulSoup(response.text, 'html.parser')
type(html_soup)
```

```python
product_list_name = html_soup.find_all(string="crossSellingProducts")
print(type(product_list_name))
print(len(product_list_name))
```
