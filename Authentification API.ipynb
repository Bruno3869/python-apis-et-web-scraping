# Authentification sur l'API de github

```python
import requests
```

```python
# Création dictionnaire contenant le token
headers = {"Authorization":"token "}
```

```python
# Requête GET
response = requests.get("https://api.github.com/users/Bruno3869", headers=headers)
```

```python
print(response.json())
```

## Autres points d'accès

```python
#huandu
response = requests.get("https://api.github.com/users/huandu", headers=headers)
huandu = response.json()
print(huandu)
```

```python
response = requests.get("https://api.github.com/orgs/facebook", headers=headers)
orgs = response.json()
print(orgs)
```
