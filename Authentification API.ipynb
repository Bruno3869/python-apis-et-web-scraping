## Requête Post

```python
# payLoad = {"name":"test", "description":"ceci est la description du repository"}
# requests.post("https://api.github.com/user/repos", json=payLoad)

payload = {"name":"api-scraping"}
response = requests.post("https://api.github.com/user/repos", json=payload, headers=headers)
statuts = response.status_code
print(statuts)
```

## Requête Patch/Put

```python
payload = {"name":"api", "description":"Super formation !"}
response = requests.patch("https://api.github.com/repos/Bruno3869/api-scraping", json=payload, headers=headers)
statuts = response.status_code
print(statuts)
```

## Requête Delete

```python
response = requests.delete("https://api.github.com/repos/Bruno3869/api", headers=headers)
statuts = response.status_code
print(statuts)
```
